```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from scipy.stats import kurtosis
from scipy.stats import skew

from IPython.display import display   # A notebook function to display more complex data (like tables)

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('healthcare-dataset-stroke-data.csv')
```

```python
df_imp =df.iloc[:,1:12]
columns_list = list(df_imp.columns)
```

```python
dfSkew = pd.DataFrame(df['avg_glucose_level']) # Create a pandas DataFrame out of the numpy array
dfSkew.plot.hist(alpha=0.5, bins=15, grid=True, legend=None)  # Pandas helper function to plot a hist. Uses matplotlib under the hood.
plt.xlabel("Average Glocuse Level")
plt.title("Skewness - Glucose Level")
plt.show()
```

```python
skew(df['avg_glucose_level'], axis = 0, bias = True)
```

```python
X = df['bmi']
X.kurt()
```

```python
plt.figure(figsize = (10,8))
sns.distplot(df.age);
```

```python
# age : When age is less than 40, chances of stroke are less. As age progresses risk increases 
# glucose level: Chances of stroke is high if glucose level is more than 150.
num_list = ['age', 'avg_glucose_level']
plt.figure(figsize=(10,4))
for i in range(len(num_list)):
    plt.subplot(1,3,i+1)
    sns.histplot(data=df_imp,x=num_list[i], hue='stroke', kde=True, stat="probability", common_norm=False)
    plt.title('Stroke vs No Stroke by {}'.format(num_list[i]))
plt.tight_layout()
```
